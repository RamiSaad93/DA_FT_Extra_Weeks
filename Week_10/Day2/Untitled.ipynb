```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 2, 3, 4, 5, 6, 7],
    'B': [5, 6, 7, 8, 5, 7, 7, 8],
    'C': [10, 10, 20, 30, 20, 20, 30, 40]
}
df = pd.DataFrame(data)

# Loop through each column in the DataFrame
for column in df.columns:
    plt.figure()  # Create a new figure for each column
    plt.scatter(df[column], bins=10, edgecolor='k', alpha=0.7)  # Plot histogram
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Sample DataFrame
data = {
    'A': [1, 2, 2, 3, 4, 5, 6, 7],
    'B': [5, 6, 7, 8, 5, 7, 7, 8],
    'C': [10, 10, 20, 30, 20, 20, 30, 40]
}
df = pd.DataFrame(data)

# Loop through each column in the DataFrame
for column in df.columns:
    # Generate the Q-Q plot data
    (quantiles, values), (slope, intercept, r) = stats.probplot(df[column], dist="norm")
    
    # Create scatter plot for Q-Q plot data
    plt.figure()  # Create a new figure for each column
    plt.scatter(quantiles, values, edgecolor='k', alpha=0.7)
    plt.plot(quantiles, slope*quantiles + intercept, 'r--')  # Reference line
    plt.title(f'Scatter Plot of Q-Q Data for {column}')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Ordered Values')
    plt.grid(True)
    plt.show()
```
